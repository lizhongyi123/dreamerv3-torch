```python
class Every:
    def __init__(self, every):
        self._every = every
        self._last = None

    def __call__(self, step):
        if not self._every:
            return 0
        if self._last is None:
            self._last = step
            return 1
        count = int((step - self._last) / self._every)
        self._last += self._every * count
        return count


```

```python
total_steps = 50
train_every = Every(5)

for i in range(50):
    print( train_every(i) )

```

```python
import torch
import torch.distributions as torchd

logits = torch.tensor([2.0, 0.5, -1.0])
dist = torchd.OneHotCategorical(logits=logits)

sample = dist.sample()
logp = dist.log_prob(sample)

print("sample:", sample)
print("log_prob:", logp)
```

```python
import torch
import torch.distributions as torchd

logits = torch.tensor([
    [2.0, 0.5, -1.0],
    [0.1, 1.2, 0.3]
])

dist = torchd.OneHotCategorical(logits=logits)
sample = dist.sample()
logp = dist.log_prob(sample)
print(sample)
print(logp)
```

```python
joint = torchd.Independent(dist, 1)
print(joint.entropy())
sample = joint.sample()
logp = joint.log_prob(sample)
print(logp)
```

```python
import torch

logits = torch.tensor([
    [2.0, 0.5, -1.0],
    [0.1, 1.2, 0.3]
])
soft = torch.softmax(logits, dim=-1)
log_soft = torch.log(soft)

print("softmax:\n", soft)
print("\nlog_softmax:\n", log_soft)

```
